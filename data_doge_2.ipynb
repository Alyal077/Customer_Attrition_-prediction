```python
numerical_cols = ['ServiceSpan','QuarterlyPayment', 'GrandPayment']
object_cols = ['sex', 'Aged', 'Married', 'TotalDependents','MobileService', '4GService', 'CyberProtection', 'HardwareSupport',
       'TechnicalAssistance', 'FilmSubscription', 'SettlementProcess']
```

```python
df_test=pd.read_csv('/content/drive/MyDrive/Hackathon/IIT KGP/test.csv')
df_test
```

```python
df_train.dropna(  axis=0, how='any', thresh=None, subset=None, inplace=True)
```

```python
df_train=df_train.drop(labels=['ID'],axis=1)
```

```python
df_train["CustomerAttrition"]=pd.factorize(df_train.CustomerAttrition)[0]
```

```python
train, test = train_test_split(df_train, test_size=0.2,shuffle=True,random_state=56)
test.head(5)
```

```python
a=train.drop(labels='CustomerAttrition',axis=1)
xtrain=a 
ytrain=train['CustomerAttrition']
b =test.drop(labels='CustomerAttrition',axis=1)
xtest=b 
ytest=test['CustomerAttrition']
```

```python
#!pip install catboost
from catboost import CatBoostClassifier
```

```python
from xgboost import XGBRegressor
xgb = XGBClassifier()
xgb.fit(xtrain,ytrain )

train_score=score(ytrain,xgb.predict(xtrain))

test_score=score(ytest,xgb.predict(xtest))

print(train_score)

print(test_score)
```

```python
CBC= CatBoostClassifier(cat_features=object_cols, scale_pos_weight=2.7666666666666666 )
CBC.fit(xtrain,ytrain )

train_score=score(ytrain,CBC.predict(xtrain))

test_score=score(ytest,CBC.predict(xtest))

print(train_score)

print(test_score)
```
